```python
groupedTract_count = points_in_map_short.groupby(['COUNTYFP','TRACTCE','ALAND']).size().reset_index(name='treesStats')
#df2 = df.rename({'a': 'X', 'b': 'Y'}, axis=1) 
groupedTract_count
```

```python
groupedTract_mean = groupedTract_count["treesStats"].mean()
print("mean",groupedTract_mean)
groupedTract_min = groupedTract_count["treesStats"].min()
print("min", groupedTract_min)
groupedTract_max = groupedTract_count["treesStats"].max()
print("max", groupedTract_max)
```

```python
import numpy as np
groupedTract_count["trees"] = np.where(groupedTract_count.treesStats < groupedTract_mean, "low", "high")
groupedTract_count[1:6]
```

```python
## remove any NTAs that are NaNs df_ML_NYC_with_trees = df_ML_NYC_with_trees.dropna() ## 2123 to 2115 so removed 8 NaN rows
df_ML_NYC_with_treesCrime_withNTAsNoNaNs = df_ML_NYC_with_treesCrime_withNTAs
df_ML_NYC_with_treesCrime_withNTAsNoNaNs[1:2]
```

```python
## find total trees for NTA areas
groupedTreesPerNTA = df_ML_NYC_with_treesCrime_withNTAsNoNaNs.groupby(["NTA Name"])["treesStats"].sum().sort_values(ascending=False).reset_index(name ='TotalNTATrees')
groupedTreesPerNTA[1:20]
```

```python
## find total crimes for NTA areas
groupedCrimesPerNTA = df_ML_NYC_with_treesCrime_withNTAsNoNaNs.groupby(["NTA Name"])["crimeStats"].sum().sort_values(ascending=False).reset_index(name ='TotalNTACrimes')
groupedCrimesPerNTA[1:200]
```

```python
df_ML_NYC_withTotalTreesPerNTA = pd.merge(df_ML_NYC_with_treesCrime_withNTAsNoNaNs, groupedTreesPerNTA, how = 'left', left_on = ['NTA Name'], right_on = ['NTA Name'])  
df_ML_NYC_withTotalTreesPerNTA[1:40]
```

```python
## find total crimes per county/tract combination
groupedCrimesPerTract = df_ML_NYC_with_treesCrime_withNTAsNoNaNs.groupby(["COUNTYFP","TRACTCE"])["crimeStats"].sum().sort_values(ascending=False).reset_index(name ='TotalTractCrimes')
groupedCrimesPerTract[1:5]
```

```python
df_ML_NYC_withTotalTreesandCrimesPerNTA = pd.merge(df_ML_NYC_withTotalTreesPerNTA, groupedCrimesPerNTA, how = 'left', left_on = ['NTA Name'], right_on = ['NTA Name'])  
df_ML_NYC_withTotalTreesandCrimesPerNTA[1:2]
```

```python
## find total crimes for county/tract areas - see 148 above - seems to be a duplicate
groupedCrimesPerTract = df_ML_NYC_with_treesCrime_withNTAsNoNaNs.groupby(["COUNTYFP","TRACTCE"])["crimeStats"].sum().sort_values(ascending=False).reset_index(name ='TotalTractCrimes')
groupedCrimesPerTract[1:5]
```

```python
df_ML_NYC_withTotalTreesandCrimesPerTract = pd.merge(df_ML_NYC_withTotalTreesandCrimesPerNTA, groupedCrimesPerTract, how = 'inner',  on = ['COUNTYFP', 'TRACTCE'])
df_ML_NYC_withTotalTreesandCrimesPerTract[1:2]
```
